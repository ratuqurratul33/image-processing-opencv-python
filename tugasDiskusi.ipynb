## Diskusi

### 1. Apa perbedaan gambar asli, grayscale, dan hasil crop?

Gambar asli (RGB) memiliki tiga channel warna yaitu merah, hijau, dan biru sehingga tampil berwarna. 
Gambar grayscale hanya memiliki satu channel yaitu tingkat keabuan sehingga tampil hitam putih. 
Sedangkan hasil crop adalah bagian tertentu dari gambar asli sehingga ukurannya lebih kecil dan hanya menampilkan area tertentu.

---

### 2. Bagaimana histogram memberikan informasi distribusi warna?

Histogram menunjukkan distribusi intensitas piksel dalam gambar. 
Sumbu X menunjukkan nilai intensitas (0–255), sedangkan sumbu Y menunjukkan jumlah piksel. 
Dengan histogram, kita dapat mengetahui apakah gambar terang, gelap, atau memiliki dominasi warna tertentu.
