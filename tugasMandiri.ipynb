# Tugas Mandiri
Pengolahan Citra dengan OpenCV

```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('pemandanganSample.jpg')

if image is None:
    print("Gambar tidak ditemukan!")
```

```python
# Resize
resize = cv2.resize(image, (300, 300))

# Crop tengah
h, w, _ = image.shape
crop = image[h//4:3*h//4, w//4:3*w//4]

# Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Histogram grayscale
hist = cv2.calcHist([gray], [0], None, [256], [0,256])

# Tampilkan hasil
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(resize, cv2.COLOR_BGR2RGB))
plt.title('Resize')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title('Crop')

plt.subplot(1,3,3)
plt.plot(hist, color='black')
plt.title('Histogram Grayscale')

plt.show()
```

## Kesimpulan Tugas MAndiri

Dari percobaan ini, gambar dapat dimanipulasi menggunakan OpenCV seperti resize dan crop. 
Histogram grayscale membantu memahami distribusi intensitas piksel dalam gambar.
